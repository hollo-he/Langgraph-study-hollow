# Test  测试
`uv add pytest`
## Getting started  入门
由于许多 LangGraph 代理依赖于状态，因此一个有用的模式是在每次使用该图的测试之前创建图，然后在测试中使用新的检查点实例编译它。
## Testing individual nodes and edges 测试单个节点和边
编译后的 LangGraph 代理会将每个节点的引用公开为 graph.nodes 。您可以利用这一点在代理中测试各个节点。请注意，这将绕过编译图时传递的任何检查点
## Partial execution  部分执行
对于由大型图组成的代理，您可能希望测试代理内部的部分执行路径，而不是端到端的整个流程。在某些情况下，将这些部分重构为子图可能更符合语义，您可以像往常一样单独调用这些子图 。
但是，如果您不想更改代理图的整体结构，可以使用 LangGraph 的持久化机制来模拟代理在所需部分开始之前暂停，并在所需部分结束时再次暂停的状态。步骤如下：
1. 使用检查点编译您的代理（内存检查点 InMemorySaver 足以满足测试需求）。
2. 调用代理的 update_state 方法，并将 as_node 参数设置为要开始测试的节点之前的节点的名称。
3. 使用与更新状态时相同的 thread_id 调用代理，并将 interrupt_after 参数设置为要停止的节点的名称。
